### <font color = green>Scaling Vectors</font>
### Scaling is growing or shrinking a vector’s length. You can grow/shrink a vector by multiplying or scaling it with a single value, known as a scalar.An important detail to note here is that scaling a vector does not change its direction, only its magnitude. When you multiply a vector by a negative number, it flips the direction of the vector .</br>

```python
v = np.array([3,1])
print(2.0 * v)
```

### we can combine two vectors and scale them to create any resulting vector we want. This whole space of possible vectors is called span, and in most cases our span can create unlimited vectors off those two vectors, simply by scaling and summing them. When we have two vectors in two different directions, they are<font color=green> linearly independent</font> and have this unlimited span.
### What happens when two vectors exist in the same direction, or exist on the same line? The combination of those vectors is also stuck on the same line, limiting our span to just that line. No matter how you scale it, the resulting sum vector is also stuck on that same line. This makes them <font color=green>linearly dependent</font>.The span here is stuck on the same line as the two vectors it is made out of. In three or more dimensions, when we have a linearly dependent set of vectors, we often get stuck on a plane in a smaller number of dimensions.
###  but why do we care whether two vectors are linearly dependent or independent? A lot of problems become difficult or unsolvable when they are linearly dependent

## <font color = red> Linear Transformations</font>

### we have two simple vectors ˆi and ˆj (“i-hat” and “j-hat”). These are known as <font color = green>basis vectors</font>, which are used to describe transformations on other vectors. They typically have a length of 1. Think of the basis vectors as building blocks to build or transform any vector. Our basis vector is expressed in a 2 × 2 matrix, where the first column is ˆi and the second column is ˆj. We can use ˆi and ˆj to create any vector we want by scaling and adding them.



### We can transform a vector with stretching, squishing, sheering, or rotating by tracking basis vector movements. This is known as a <font color=green>linear transformation</font>.




### ### A <font color=green>matrix</font> is a collection of vectors (such as ˆi, ˆj) that can have multiple rows and columns and is a convenient way to package data. a matrix really is a transformation expressed as basis vectors.


```python
basis = np.array([[2,0], [0,3]])
print(basis,'\n*********')
v = np.array([2,1])
print(v,'\n*********')
result = basis.dot(v)
print(result)                      
```

### You might be wondering how we think of vector transformations in 3d or more. The concept of basis vectors extends quite nicely. If I have a three-dimensional vector space, then I have basis vectors ˆi, ˆj, and ˆk

## <font color=red> Matrix Multiplication</font>
### We learned how to multiply a vector and a matrix, but what exactly does multiplying two matrices accomplish? Think of matrix multiplication as applying multiple transformations to a vector space. You multiply and add each row from the first matrix to each respective column of the second matrix.


```python
m1 = np.array([[2,3],[4,2]])
m2 = np.array([[1,4],[2,5]])
print(m1,'\n*********\n',m2,'\n*********')
result = np.matmul(m1,m2) #result = m1.dot(m2) #result = m1 @ m2
print(result)
# remember that np.matmul(m1,m2) and np.matmul(m2,m1) produces different results!
```

## <font color=red>Determinants</font>
###  Determinants describe how much a sampled area in a vector space changes in scale with linear transformations. The determinant is a special number that can be calculated from a matrix and matrix should be square (equal rows and cols). The determinant helps us find the inverse of a matrix, tells us things about the matrix that are useful in systems of linear equations, calculus and more.</br> |A| = ad − bc</br> |A| = a(ei − fh) − b(di − fg) + c(dh − eg) 

```python
from numpy.linalg import det
m = np.array([[2,3,5],[4,1,2],[1,1,3]])
print(m,'\n*********')
print(det(m))
```

## <font color=red>Special Types of Matrices</font>

### <font color=green>Square Matrix</font> : The square matrix is a matrix that has an equal number of rows and columns. 
### <font color=green>Identity Matrix</font> : The identity matrix is a square matrix that has a diagonal of 1s while the other values are 0.
### <font color=green>Inverse Matrix</font> : The inverse of matrix A is called A^ -1. When we perform matrix multiplication between A^−1 and A, we end up with an identity matrix. 

```python
from numpy.linalg import inv
m = np.array([[4,2,4],[5,3,7],[9,3,6]])
print(m,'\n*********')
print(inv(m))
```

### <font color=green>Diagonal Matrix</font> : Similar to the identity matrix is the diagonal matrix, which has a diagonal of nonzero values while the rest of the values are 0.
### <font color=green>Triangular Matrix</font> : Similar to the diagonal matrix is the triangular matrix, which has a diagonal of nonzero values in front of a triangle of values, while the rest of the values are 0.
### <font color=green>Sparse Matrix</font> : Occasionally, you will run into matrices that are mostly zeroes and have very few nonzero elements. These are called sparse matrices.

## <font color=red>Systems of Equations and Inverse Matrices</font>
### One of the basic use cases for linear algebra is solving systems of equations. </br>  </br>If a computer wants to solve it, you will need to express this problem in terms of matrices.</br> </br> 

```python
from numpy.linalg import inv

A = np.array([ [4, 2, 4],[5, 3, 7],[9, 3, 6] ])
B = np.array([44,56,72])
X = inv(A).dot(B)

print(X)
print("x : ",X[0]," ,y : ",X[1]," ,z : ",X[2])
```

## <font color=red>Eigenvectors and Eigenvalues</font>
### Matrix decomposition is helpful for tasks like finding inverse matrices and calculating determinants, as well as linear regression.</br> </br>Let’s focus on a common method called <font color=green>eigendecomposition</font>, which is often used for machine learning and principal component analysis. eigendecomposition is helpful for breaking up a matrix into components that are easier to work with in different machine learning tasks.It only works on square matrices.</br> </br>In eigendecomposition, there are two components: the eigenvalues denoted by lambda λ and eigenvector by v. </br>If we have a square matrix A, it has the following eigenvalue equation: <font color=green>Av = λv</font></br> </br>There is one eigenvector and eigenvalue for each dimension of the parent matrix, and not all matrices can be decomposed into an eigenvector and eigenvalue. Sometimes complex (imaginary) numbers will even result.</br><font color=green>A = Q Λ Q^−1</font> In this new formula, Q is the eigenvectors, Λ is the eigenvalues in diagonal form, and Q^−1 is the inverse matrix of Q.




```python
from numpy import diag
from numpy.linalg import eig, inv

A = np.array([[1, 2],[4, 5]]); print(A,'\n*********')

eigenvals, eigenvecs = eig(A)
print("EIGENVALUES : \n",eigenvals,'\n********'); print("EIGENVECTORS : \n",eigenvecs,'\n********')

Q = eigenvecs; L = diag(eigenvals); Qinv = inv(Q)

#remember the order of variables in the multiplication
res = Q @ L @ Qinv
print(res)
```
